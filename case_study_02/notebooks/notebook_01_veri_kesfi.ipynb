# Faz 1: Veri Kesfi ve Tanimlayici Istatistik

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os

warnings.filterwarnings("ignore")
plt.rcParams["figure.dpi"] = 150

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.float_format", lambda x: "%.3f" % x)

DATA_DIR = "../data"
FIGURES_DIR = "../outputs/figures"
os.makedirs(FIGURES_DIR, exist_ok=True)

# Pickle dosyalarindan oku
df_tahsilat = pd.read_pickle(os.path.join(DATA_DIR, "tahsilat.pkl"))
df_tahsilat_1 = pd.read_pickle(os.path.join(DATA_DIR, "tahsilat_1.pkl"))
df_tahakkuk = pd.read_pickle(os.path.join(DATA_DIR, "tahakkuk.pkl"))
df_tahakkuk_1 = pd.read_pickle(os.path.join(DATA_DIR, "tahakkuk_1.pkl"))
df_tahakkuk_2 = pd.read_pickle(os.path.join(DATA_DIR, "tahakkuk_2.pkl"))

print("Veri Setleri:")
for name, df in [("df_tahsilat", df_tahsilat), ("df_tahsilat_1", df_tahsilat_1),
                 ("df_tahakkuk", df_tahakkuk), ("df_tahakkuk_1", df_tahakkuk_1),
                 ("df_tahakkuk_2", df_tahakkuk_2)]:
    print(f"  {name}: {df.shape}")
```

```python
# 1.1 Veri Yapisi Inceleme - Tahsilat
print(f"Tahsilat: {df_tahsilat.shape}")
df_tahsilat.info()
display(df_tahsilat.head(3))
display(df_tahsilat.describe())
```

```python
# 1.1 Veri Yapisi Inceleme - Tahsilat 1
print(f"Tahsilat 1: {df_tahsilat_1.shape}")
df_tahsilat_1.info()
display(df_tahsilat_1.head(3))
display(df_tahsilat_1.describe())
```

```python
# 1.1 Veri Yapisi Inceleme - Tahakkuk (3 ilce)
for df_tmp, ilce in [(df_tahakkuk, "Hamamozu"), (df_tahakkuk_1, "Gumushaciköy"), (df_tahakkuk_2, "Göynücek")]:
    print(f"\nTahakkuk - {ilce}: {df_tmp.shape}")
    display(df_tmp.head(3))
    display(df_tmp.describe())
```

```python
# 1.2 Benzersiz Musteri Sayilari
musteri_hamamozu = df_tahakkuk["sozlesme_hesap_no"].nunique()
musteri_gumushaciköy = df_tahakkuk_1["sozlesme_hesap_no"].nunique()
musteri_goynucek = df_tahakkuk_2["sozlesme_hesap_no"].nunique()

musteri_df = pd.DataFrame({
    "Ilce": ["Hamamozu", "Gumushaciköy", "Göynücek", "TOPLAM"],
    "Benzersiz Musteri": [musteri_hamamozu, musteri_gumushaciköy, musteri_goynucek,
                          musteri_hamamozu + musteri_gumushaciköy + musteri_goynucek],
    "Kayit Sayisi": [len(df_tahakkuk), len(df_tahakkuk_1), len(df_tahakkuk_2),
                     len(df_tahakkuk) + len(df_tahakkuk_1) + len(df_tahakkuk_2)]
})
musteri_df["Musteri Basina Ort. Kayit"] = (musteri_df["Kayit Sayisi"] / musteri_df["Benzersiz Musteri"]).round(1)
display(musteri_df)
```

```python
# 1.3 Tahakkuk Verilerini Birlestirme
df_tahakkuk_all = pd.concat([df_tahakkuk, df_tahakkuk_1, df_tahakkuk_2], ignore_index=True)
print(f"Birlestirilmis Tahakkuk: {df_tahakkuk_all.shape}")
display(df_tahakkuk_all["ilce"].value_counts())
```

```python
# 1.4 Veri Kalitesi Kontrolu (kwh sutunu)
kwh = df_tahakkuk_all["kwh"]
n = len(df_tahakkuk_all)

eksik = kwh.isna().sum()
negatif = (kwh < 0).sum()
sifir = (kwh == 0).sum()

Q1, Q3 = kwh.quantile(0.25), kwh.quantile(0.75)
IQR = Q3 - Q1
outlier_count = ((kwh < Q1 - 1.5 * IQR) | (kwh > Q3 + 1.5 * IQR)).sum()

kalite_df = pd.DataFrame({
    "Metrik": ["Eksik deger", "Negatif deger", "Sifir deger", "Outlier (IQR)"],
    "Sayi": [eksik, negatif, sifir, outlier_count],
    "Oran (%)": [eksik/n*100, negatif/n*100, sifir/n*100, outlier_count/n*100]
}).round(2)
display(kalite_df)

print(f"\nIQR: Q1={Q1:.2f}, Q3={Q3:.2f}, IQR={IQR:.2f}")
print(f"Alt sinir: {Q1 - 1.5*IQR:.2f} | Ust sinir: {Q3 + 1.5*IQR:.2f}")
print(f"\nTemel istatistikler:")
display(kwh.describe())

# Negatif degerlerin dagilimi
if negatif > 0:
    negatif_df = df_tahakkuk_all[df_tahakkuk_all["kwh"] < 0]
    print(f"\nNegatif degerlerin ilce dagilimi:")
    display(negatif_df["ilce"].value_counts())
    print(f"Negatif degerlerin hesap sinifi dagilimi (ilk 10):")
    display(negatif_df["Hesap Sınıfı"].value_counts().head(10))
```

```python
# 1.5 Hesap Sinifina Göre Tuketim Istatistikleri
hesap_istatistik = df_tahakkuk_all.groupby("Hesap Sınıfı")["kwh"].agg(
    Kayit="count", Ortalama="mean", Medyan="median", Std="std", Min="min", Max="max"
).round(2)
hesap_istatistik["Oran (%)"] = (hesap_istatistik["Kayit"] / len(df_tahakkuk_all) * 100).round(2)
hesap_istatistik = hesap_istatistik.sort_values("Kayit", ascending=False)
display(hesap_istatistik)
```
